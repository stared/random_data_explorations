```python
import pandas as pd
%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
import datetime
sns.set_style('whitegrid')
```

```python
# 2018
# male      325
# female     93
```

```python
gender = pd.DataFrame({"male": [271, 325], "female": [56, 93]}, index=[2017, 2018])
gender.plot.bar(stacked=True)  # stacked=True
plt.ylabel("#participants")
plt.title("PyData Warsaw Conference")
```

```python
perc_women = 100 * pd.Series([56/415, 93/435], index=[2017, 2018])
perc_women.plot.bar()  # stacked=True
```

```python
# Poland 2017 was: 332 / 415 = 0.8
# Poland 2018 is: 0.65
```

```python
# 2017
# Warsaw 181 / 415
# Poland 332 + 4 / 415

# 2018
# Warsaw 163 / 435
# Poland 284 + 17 / 435

location = pd.DataFrame({"Warsaw": [181, 163],
                         "Poland - Warsaw": [332 + 4 - 181, 284 + 17 - 163],
                         "international": [415 - (332 + 4), 435 - (284 + 17)]},
                        index=[2017, 2018])[['Warsaw', 'Poland - Warsaw', 'international']]
location.plot.bar()  # stacked=True
plt.ylabel("#participants")
plt.title("PyData Warsaw Conference: where did you come from?")
plt.legend(bbox_to_anchor=(1., 0.95))
```

```python
perc_various = 100 * pd.DataFrame({"international": [1 - (332 + 4)/415, 1 - (284 + 17)/435], "women": [56/415, 93/435]}, index=[2017, 2018])
perc_various[['women', 'international']].plot.bar()
plt.ylabel("% of participants")
plt.title("PyData Warsaw Conference: participant diversity")
plt.savefig("")
```

Moreover, the number of countries went up from 19 to 26.

```python
# countries (including Poland minus unknown)
# 2017
# 19

# 2018
# 26
```

```python
countries_count = pd.Series([19, 26], index=[2017, 2018])
countries_count.plot.bar()  # stacked=True
```
