```python
# https://stackoverflow.com/questions/40450943/adding-labels-to-a-bokeh-plot
from bokeh.io import output_notebook
from bokeh.plotting import show, ColumnDataSource
from bokeh.charts import Scatter
from bokeh.models import LabelSet
from pandas.core.frame import DataFrame

output_notebook()

source = DataFrame(
    dict(
        off_rating=[66, 71, 72, 68, 58, 62],
        def_rating=[165, 189, 220, 141, 260, 174],
        names=['Mark', 'Amir', 'Matt', 'Greg', 'Owen', 'Juan']
    )
)


scatter_plot = Scatter(
                source,
                x='off_rating',
                y='def_rating',
                title='Offensive vs. Defensive Eff',
                color='navy')

labels = LabelSet(
            x='off_rating',
            y='def_rating',
            text='names',
            level='glyph',
            x_offset=5, 
            y_offset=5, 
            source=ColumnDataSource(source), 
            render_mode='canvas')

scatter_plot.add_layout(labels)

show(scatter_plot)
```
